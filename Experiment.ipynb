```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Person": list(range(1, 16)),
    "Height cm": [160, 165, 158, 170, 172, 168, 175, 162, 160, 168, 173, 165, 166, 164, 176],
    "Weight kg": [45, 50, 52, 60, 65, 68, 72, 74, 77, 80, 85, 58, 62, 69, 90],
    "Category": ['Slim', 'slim', 'slim', 'fit', 'fit', 'fit', 'fit', 'fit', 
                 'obese', 'obese', 'fit', 'fit', 'fit', 'fit', 'fit']
}

df = pd.DataFrame(data)
df.columns = df.columns.str.strip()
df['Category'] = df['Category'].str.capitalize()

plt.figure(figsize=(10, 4))
sns.scatterplot(x=df['Weight kg'], y=[0] * len(df), hue=df['Category'], s=100)
plt.yticks([])
plt.xlabel("Weight (kg)")
plt.title("Univariate Analysis of Weight by Category")
plt.legend(title="Category")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Height cm', y='Weight kg', hue='Category', s=100)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Bivariate Analysis of Height and Weight by Category")
plt.legend(title="Category")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['Age'] = [25, 30, 22, 35, 40, 32, 45, 50, 55, 60, 65, 28, 33, 26, 70]

sns.pairplot(df, hue='Category', vars=['Height cm', 'Weight kg', 'Age'])

plt.suptitle("Multivariate Analysis of Height, Weight, and Age by Category", y=1.02)

plt.show()

```
